```python
import pandas as pd
```

```python
df_rho = pd.read_csv('../output/Rho统计时间.tsv',sep='\t')
df_number = pd.read_csv('../output/Rho_number.csv')
```

```python
df_rho = df_rho.merge(df_number,on='Haplogroup',how='left')
df_rho.fillna(0,inplace=True)
df_rho.loc[df_rho['Count'].astype(int)>=10]
df_rho.drop(columns=['RHO_SYN_AGE','RHO_SYN_AGE_95LB','RHO_SYN_AGE_95HB'],inplace=True)
df_rho.drop_duplicates(subset=['Haplogroup'],inplace=True)
```

```python
df_rho = df_rho.loc[df_rho['Count']>=10]
df_rho.iloc[:, 1:] = df_rho.iloc[:,1:].clip(lower=0)
df_rho.to_csv('../output/Rho统计时间_trim.csv',index=False)
```
